```julia
using Plots

X = [1; 2; -0.5; -2; 2]
Y = [2; 3.35631; 0.76225; 0.126; 4.2563]

x = -3:0.01:3
```

```julia
A = [X' * X sum(X); sum(X) length(X)];
b = [X' * Y; sum(Y)];
Coeff1 = A \ b;

display(Coeff1)

p1(x) = @. Coeff1[1]*x+Coeff1[2]
plot(X, p1(X))
scatter!(X, Y)
```

```julia
Y_ln = log.(Y)

A = [X' * X sum(X); sum(X) length(X)];
b = [X' * Y_ln; sum(Y_ln)];

Coeff2 = A \ b;

display(Coeff2)

p2(x) = @. exp(Coeff2[2] + Coeff2[1] * x)

Plots.plot(x,p2(x),legend=false)
Plots.scatter!(X,Y)
```

```julia
plot(x, p1(x), label="Lin function")
plot!(x, p2(x), label="Exp function")
scatter!(X, Y, label="(X, Y)")
```

```julia
lin_err = abs.(p1(X) .- Y)
exp_err = abs.(p2(X) .- Y)

println("Max Lin Error: ", maximum(lin_err))
println("Tot Lin Error: ", sum(lin_err))
println("Max Exp Error: ", maximum(exp_err))
println("Tot Exp Error: ", sum(exp_err))
println("Exponential interpolating function has the highest single error")
println("Linear interpolating function has the larger cummulative error")
```
